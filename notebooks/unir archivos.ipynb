```python
import pandas as pd
import glob
import os

# Directorio donde están los CSV
directorio = r"C:\DATOS\CSV"

# Obtener lista de archivos
archivos = glob.glob(os.path.join(directorio, "*.CSV"))

# Leer y unir
df_final = pd.concat(
    [pd.read_csv(archivo, encoding="latin-1", skiprows= 3) for archivo in archivos],
    ignore_index=True
)

# Guardar resultado
df_final.to_csv(
    os.path.join(directorio, "archivo_unido.csv"),
    index=False,
    encoding="utf-8"
)

print(f"Se unieron {len(archivos)} archivos")
print(f"Total registros: {len(df_final)}")
```
